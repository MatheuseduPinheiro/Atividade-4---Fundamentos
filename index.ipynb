## A2. Regress√£o Linear usando linregress

#### A fun√ß√£o linregress do SciPy realiza uma regress√£o linear e retorna

### üìà Slope (Coeficiente Angular)

**slope** (tamb√©m conhecido como **coeficiente angular** ou **coeficiente de inclina√ß√£o**) representa a varia√ß√£o m√©dia na vari√°vel dependente (**Y**) para cada unidade de varia√ß√£o na vari√°vel independente (**X**). 

√â a medida da **taxa de mudan√ßa** da vari√°vel dependente em rela√ß√£o √† vari√°vel independente.

---

#### üßÆ F√≥rmula:

A f√≥rmula para calcular o **slope** (`b‚ÇÅ`) na regress√£o linear simples √©:

$$
b_1 = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sum (x_i - \bar{x})^2}
$$

**Onde:**

- $x_i$: valor da vari√°vel independente (X) para a observa√ß√£o *i*  
- $y_i$: valor da vari√°vel dependente (Y) para a observa√ß√£o *i*  
- $\bar{x}$: m√©dia da sequ√™ncia amostral independente (X)  
- $\bar{y}$: m√©dia da sequ√™ncia amostral dependente (Y)

### üìç Intercepto (Constante da Regress√£o)

O **intercepto**, tamb√©m conhecido como **intercepto y** ou **constante**, representa o **valor m√©dio da vari√°vel dependente (Y)** quando todas as vari√°veis independentes (X) s√£o iguais a zero.

Em termos mais simples, √© o **ponto onde a linha de regress√£o intercepta o eixo Y** no gr√°fico.

---

#### üßÆ F√≥rmula:

A equa√ß√£o geral da regress√£o linear simples √©:

$$
Y = a + bX
$$

**Onde:**

- **Y**: vari√°vel dependente (a vari√°vel que estamos tentando prever)  
- **a**: intercepto (valor de Y quando X = 0)  
- **b**: inclina√ß√£o da linha de regress√£o (a mudan√ßa em Y para cada unidade de mudan√ßa em X)  
- **X**: vari√°vel independente (a vari√°vel que usamos para prever Y)


### üìä Coeficiente de Correla√ß√£o (r)

O **coeficiente de correla√ß√£o**, geralmente representado por **"r"**, indica a **for√ßa** e a **dire√ß√£o** da rela√ß√£o linear entre duas vari√°veis. Ele varia entre -1 e 1:

- **r = 1**: Correla√ß√£o linear positiva perfeita (se uma vari√°vel aumenta, a outra tamb√©m aumenta na mesma propor√ß√£o)  
- **r = -1**: Correla√ß√£o linear negativa perfeita (se uma vari√°vel aumenta, a outra diminui na mesma propor√ß√£o)  
- **r = 0**: N√£o h√° correla√ß√£o linear entre as vari√°veis

---

#### üßÆ F√≥rmula (Correla√ß√£o de Pearson):

$$
r = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{n \cdot s_x \cdot s_y}
$$

**Onde:**

- \( x_i \): valor da vari√°vel x na observa√ß√£o *i*  
- \( y_i \): valor da vari√°vel y na observa√ß√£o *i*  
- \( \bar{x} \): m√©dia da vari√°vel x  
- \( \bar{y} \): m√©dia da vari√°vel y  
- \( n \): n√∫mero de observa√ß√µes  
- \( s_x \): desvio padr√£o da vari√°vel x  
- \( s_y \): desvio padr√£o da vari√°vel y


### üìä p-value (Valor-p)

O **p-value** representa a **probabilidade de observar uma estat√≠stica de teste pelo menos t√£o extrema quanto a calculada a partir dos dados da amostra**, assumindo que a **hip√≥tese nula** seja verdadeira.

Ele √© usado para determinar a **signific√¢ncia estat√≠stica** dos resultados em testes de hip√≥tese.

---

#### üßÆ F√≥rmula (p-value):

$$
z = \frac{\bar{x} - \mu_0}{\frac{\sigma}{\sqrt{n}}}
$$

**Onde:**

- \( \bar{x} \): m√©dia da amostra (sample mean)  
- \( \mu_0 \): m√©dia hipot√©tica (hypothesized mean, sob a hip√≥tese nula)  
- \( \sigma \): desvio padr√£o populacional (standard deviation) 0000 
- \( n \): tamanho da amostra (sample size) 0000

obs: 

0000 :erro padr√£o associado √† amostra


### üìâ Erro Padr√£o da Estimativa (EPE)

O **Erro Padr√£o da Estimativa (EPE)** em regress√£o linear √© uma medida da dispers√£o dos pontos de dados em torno da reta de regress√£o.  

Ele indica, em m√©dia, **qu√£o distante os valores observados da vari√°vel dependente est√£o dos valores preditos pelo modelo**.  
Em outras palavras, o EPE quantifica a **precis√£o da regress√£o** na predi√ß√£o dos valores de \( Y \).

---

#### üßÆ F√≥rmula do Erro Padr√£o da Estimativa:

$$
s_e = \sqrt{\frac{\text{MSE}}{n - p - 1}}
$$

**Onde:**

- \( s_e \): Erro padr√£o da estimativa  
- **MSE**: Erro quadr√°tico m√©dio (Mean Squared Error)  
- \( n \): N√∫mero total de observa√ß√µes  
- \( p \): N√∫mero de vari√°veis independentes (regressores) no modelo, **excluindo** o intercepto

### Fun√ß√£o Densidade de Probabilidade da Distribui√ß√£o Normal

$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}} \cdot e^{ -\frac{1}{2} \left( \frac{x - \mu}{\sigma} \right)^2 }
$$

**Onde:**

- \( x \): valor da vari√°vel aleat√≥ria  
- \( \mu \): m√©dia da distribui√ß√£o (ex.: `y_m`)  
- \( \sigma \): desvio padr√£o (ex.: `desvio_padrao_amostral_Y`)  
- \( f(x) \): valor da densidade de probabilidade no ponto \( x \)


#### Distribui√ß√£o Padr√£o das Amostral

```python
def media_amostral(sequencia):
    acumulador = 0
    for i in range(len(sequencia)):
        acumulador += sequencia[i]
    return acumulador / len(sequencia)
```

```python
def desvio_padrao_amostral(sequencia, media):
    soma_diferencas_quadrado = 0
    for i in range(len(sequencia)):
        soma_diferencas_quadrado += (sequencia[i] - media) ** 2
    return (soma_diferencas_quadrado / (len(sequencia) - 1)) ** 0.5
```

```python
def p_value(media_populacao, media_amostral, variancia_populacional, tamanho_sequencia):
    desvio_padrao_populacional = np.sqrt(variancia_populacional)  # Calculando o desvio padr√£o populacional
    return (media_amostral - media_populacao) / (desvio_padrao_populacional / np.sqrt(tamanho_sequencia))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Gerando sequ√™ncia com seed
np.random.seed(42)
seq = np.random.randint(1, 100, size=25)  # Seus dados originais

# C√°lculos estat√≠sticos
seq_m = np.mean(seq)  # M√©dia amostral
desvio_padrao_amostral_seq = np.std(seq, ddof=1)  # Desvio padr√£o amostral
mu = 50  # Hip√≥tese de m√©dia populacional

# Teste t para obter o p-valor (comparando com mu=50)
t_stat, p_valor = stats.ttest_1samp(seq, mu)

# Gerando valores da curva normal te√≥rica
x_vals = np.linspace(min(seq) - 15, max(seq) + 15, 100)
pdf = (1 / (desvio_padrao_amostral_seq * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x_vals - mu) / desvio_padrao_amostral_seq) ** 2)

# Plotando
plt.figure(figsize=(12, 6))

# Histograma dos dados (Agora com len(seq) para bins)
plt.hist(seq, bins=len(seq), density=True, alpha=0.5, color='skyblue', edgecolor='black', label='Dados amostrais')

# Curva normal te√≥rica
plt.plot(x_vals, pdf, 'b-', linewidth=2, label=f'Normal (Œº={seq_m:.1f}, œÉ={desvio_padrao_amostral_seq:.1f})')

# Linhas de refer√™ncia
plt.axvline(seq_m, color='red', linestyle='--', linewidth=1.5, label=f'M√©dia amostral = {seq_m:.1f}')


# Configura√ß√µes do gr√°fico
plt.title(f'Distribui√ß√£o Amostral vs Normal Te√≥rica\n(p-valor = {p_valor:.4f})', fontsize=14, pad=20)
plt.xlabel('Valores', fontsize=12)
plt.ylabel('Densidade de Probabilidade', fontsize=12)
plt.grid(axis='y', alpha=0.3)

# Adicionando a linha vertical para o p-valor
plt.axvline(p_valor, color='purple', linestyle='-', linewidth=2, label=f'P-valor = {p_valor:.4f}')

# Legenda melhor posicionada
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.tight_layout()

plt.show()

# Mostrando estat√≠sticas
print("\nEstat√≠sticas:")
print(f"M√©dia amostral: {seq_m:.2f}")
print(f"Desvio padr√£o amostral: {desvio_padrao_amostral_seq:.2f}")
print(f"Teste t contra Œº={mu}: p-valor = {p_valor:.4f}")
print(f"Sequ√™ncia Amostrais: {seq}")
print(f"Tamanho dos Dados Amostrais: {len(seq)}")
```

### Aplica√ß√£o do P-value na Hipotese Nula

#### Criar dados simulados

```python
# Semente para reproduzibilidade
np.random.seed(42)

# Criar dados simulados
grupo_controle = np.random.normal(loc=50, scale=10, size=100)
grupo_teste = np.random.normal(loc=55, scale=10, size=100)
```

#### Formular hip√≥teses

- Hip√≥tese nula (H‚ÇÄ): N√£o h√° diferen√ßa significativa entre as m√©dias dos grupos de controle e teste (Œº_controle = Œº_teste)

- Hip√≥tese alternativa (H‚ÇÅ): H√° diferen√ßa significativa entre as m√©dias (Œº_controle ‚â† Œº_teste)

#### Realizar teste t de duas amostras

```python
# Realizar teste t
t_stat, p_value = stats.ttest_ind(grupo_controle, grupo_teste)

print(f"Estat√≠stica t: {t_stat:.4f}")
print(f"Valor-p: {p_value:.4f}")
```

#### Interpretar o valor-p

- Se p-value ‚â§ 0.05: Rejeitamos H‚ÇÄ (diferen√ßa significativa)

- Se p-value > 0.05: N√£o rejeitamos H‚ÇÄ (sem evid√™ncia de diferen√ßa)

#### Criar boxplot com as amostras

```python
plt.figure(figsize=(10, 6))

# Boxplot
plt.boxplot([grupo_controle, grupo_teste], 
            positions=[1, 2], 
            widths=0.6,
            patch_artist=True,
            boxprops=dict(facecolor='lightblue'))

# Plotar pontos individuais (amostras)
for i in range(len(grupo_controle)):
    plt.plot(1 + np.random.normal(0, 0.05), grupo_controle[i], 'o', color='blue', alpha=0.5)
for i in range(len(grupo_teste)):
    plt.plot(2 + np.random.normal(0, 0.05), grupo_teste[i], 'o', color='red', alpha=0.5)

# Adicionar t√≠tulo e labels
plt.title('Compara√ß√£o entre Grupo Controle e Grupo Teste', pad=20)
plt.xticks([1, 2], ['Controle', 'Teste'])
plt.ylabel('Valor da M√©trica')
plt.grid(axis='y', alpha=0.4)

# Adicionar informa√ß√£o do valor-p
plt.text(1.5, max(np.concatenate([grupo_controle, grupo_teste])) * 1.05,
         f'Valor-p = {p_value:.4f}', ha='center')

plt.tight_layout()
plt.show()
```

- Caixas (boxes): Mostram os quartis (25%, mediana, 75%) dos dados

- Bigodes (whiskers): Mostram o range dos dados (normalmente 1.5x o IQR)

- Pontos individuais: Mostram cada observa√ß√£o nos dados (transparentes para melhor visualiza√ß√£o)

- Posi√ß√£o: Grupo controle √† esquerda (1), grupo teste √† direita (2)

- Valor-p: Mostrado no topo do gr√°fico para refer√™ncia
