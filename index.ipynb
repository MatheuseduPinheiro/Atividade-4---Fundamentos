### 📈 Slope (Coeficiente Angular)

**slope** (também conhecido como **coeficiente angular** ou **coeficiente de inclinação**) representa a variação média na variável dependente (**Y**) para cada unidade de variação na variável independente (**X**). 

É a medida da **taxa de mudança** da variável dependente em relação à variável independente.

---

#### 🧮 Fórmula:

A fórmula para calcular o **slope** (`b₁`) na regressão linear simples é:

$$
b_1 = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sum (x_i - \bar{x})^2}
$$

**Onde:**

- $x_i$: valor da variável independente (X) para a observação *i*  
- $y_i$: valor da variável dependente (Y) para a observação *i*  
- $\bar{x}$: média da sequência amostral independente (X)  
- $\bar{y}$: média da sequência amostral dependente (Y)

### 📍 Intercepto (Constante da Regressão)

O **intercepto**, também conhecido como **intercepto y** ou **constante**, representa o **valor médio da variável dependente (Y)** quando todas as variáveis independentes (X) são iguais a zero.

Em termos mais simples, é o **ponto onde a linha de regressão intercepta o eixo Y** no gráfico.

---

#### 🧮 Fórmula:

A equação geral da regressão linear simples é:

$$
Y = a + bX
$$

**Onde:**

- **Y**: variável dependente (a variável que estamos tentando prever)  
- **a**: intercepto (valor de Y quando X = 0)  
- **b**: inclinação da linha de regressão (a mudança em Y para cada unidade de mudança em X)  
- **X**: variável independente (a variável que usamos para prever Y)


### 📊 Coeficiente de Correlação (r)

O **coeficiente de correlação**, geralmente representado por **"r"**, indica a **força** e a **direção** da relação linear entre duas variáveis. Ele varia entre -1 e 1:

- **r = 1**: Correlação linear positiva perfeita (se uma variável aumenta, a outra também aumenta na mesma proporção)  
- **r = -1**: Correlação linear negativa perfeita (se uma variável aumenta, a outra diminui na mesma proporção)  
- **r = 0**: Não há correlação linear entre as variáveis

---

#### 🧮 Fórmula (Correlação de Pearson):

$$
r = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{n \cdot s_x \cdot s_y}
$$

**Onde:**

- \( x_i \): valor da variável x na observação *i*  
- \( y_i \): valor da variável y na observação *i*  
- \( \bar{x} \): média da variável x  
- \( \bar{y} \): média da variável y  
- \( n \): número de observações  
- \( s_x \): desvio padrão da variável x  
- \( s_y \): desvio padrão da variável y


### 📊 p-value (Valor-p)

O **p-value** representa a **probabilidade de observar uma estatística de teste pelo menos tão extrema quanto a calculada a partir dos dados da amostra**, assumindo que a **hipótese nula** seja verdadeira.

Ele é usado para determinar a **significância estatística** dos resultados em testes de hipótese.

---

#### 🧮 Fórmula (p-value):

$$
z = \frac{\bar{x} - \mu_0}{\frac{\sigma}{\sqrt{n}}}
$$

**Onde:**

- \( \bar{x} \): média da amostra (sample mean)  
- \( \mu_0 \): média hipotética (hypothesized mean, sob a hipótese nula)  
- \( \sigma \): desvio padrão populacional (standard deviation) 0000 
- \( n \): tamanho da amostra (sample size) 0000

obs: 

0000 :erro padrão associado à amostra


### 📉 Erro Padrão da Estimativa (EPE)

O **Erro Padrão da Estimativa (EPE)** em regressão linear é uma medida da dispersão dos pontos de dados em torno da reta de regressão.  

Ele indica, em média, **quão distante os valores observados da variável dependente estão dos valores preditos pelo modelo**.  
Em outras palavras, o EPE quantifica a **precisão da regressão** na predição dos valores de \( Y \).

---

#### 🧮 Fórmula do Erro Padrão da Estimativa:

$$
s_e = \sqrt{\frac{\text{MSE}}{n - p - 1}}
$$

**Onde:**

- \( s_e \): Erro padrão da estimativa  
- **MSE**: Erro quadrático médio (Mean Squared Error)  
- \( n \): Número total de observações  
- \( p \): Número de variáveis independentes (regressores) no modelo, **excluindo** o intercepto

### Função Densidade de Probabilidade da Distribuição Normal

$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}} \cdot e^{ -\frac{1}{2} \left( \frac{x - \mu}{\sigma} \right)^2 }
$$

**Onde:**

- \( x \): valor da variável aleatória  
- \( \mu \): média da distribuição (ex.: `y_m`)  
- \( \sigma \): desvio padrão (ex.: `desvio_padrao_amostral_Y`)  
- \( f(x) \): valor da densidade de probabilidade no ponto \( x \)


#### Distribuição Padrão das Amostral

```python
def media_amostral(sequencia):
    acumulador = 0
    for i in range(len(sequencia)):
        acumulador += sequencia[i]
    return acumulador / len(sequencia)
```

```python
def desvio_padrao_amostral(sequencia, media):
    soma_diferencas_quadrado = 0
    for i in range(len(sequencia)):
        soma_diferencas_quadrado += (sequencia[i] - media) ** 2
    return (soma_diferencas_quadrado / (len(sequencia) - 1)) ** 0.5
```

## B - Calculando um intervalo de confiança 

```python
import scipy.stats as stats
import math

# Specify sample mean (x_bar), sample standard deviation (s), sample size (n) and confidence level
x_bar = 62.1
s = 13.46
n = 30
confidence_level = 0.95

# Calculate alpha, degrees of freedom (df), the critical t-value, and the margin of error
alpha = (1-confidence_level)
df = n - 1
standard_error = s/math.sqrt(n)
critical_t = stats.t.ppf(1-alpha/2, df)
margin_of_error = critical_t * standard_error

# Calculate the lower and upper bound of the confidence interval
lower_bound = x_bar - margin_of_error
upper_bound = x_bar + margin_of_error

# Print the results
print("Critical t-value: {:.3f}".format(critical_t))
print("Margin of Error: {:.3f}".format(margin_of_error))
print("Confidence Interval: [{:.3f},{:.3f}]".format(lower_bound,upper_bound))
print("The {:.1%} confidence interval for the population mean is:".format(confidence_level))
print("between {:.3f} and {:.3f}".format(lower_bound,upper_bound))
```
